```python
import pandas as pd
```

```python
spb_flat_1room_sale = pd.read_csv('cian_parsing_result_sale_1_100_sankt-peterburg_03_Apr_2023_1.csv', 
                                  encoding = 'windows-1251', sep = ';')
spb_flat_2rooms_sale = pd.read_csv('cian_parsing_result_sale_1_100_sankt-peterburg_04_Apr_2023_2.csv', 
                               encoding = 'windows-1251', sep = ';')
spb_flat_3rooms_sale = pd.read_csv('cian_parsing_result_sale_1_100_sankt-peterburg_04_Apr_2023_3.csv',
                               encoding = 'windows-1251', sep = ';')
spb_flat_4rooms_sale = pd.read_csv('cian_parsing_result_sale_1_100_sankt-peterburg_04_Apr_2023_4.csv',
                               encoding = 'windows-1251', sep = ';')
spb_flat_5rooms_sale = pd.read_csv('cian_parsing_result_sale_1_100_sankt-peterburg_04_Apr_2023_5.csv',
                               encoding = 'windows-1251', sep = ';')
spb_flat_studio_sale = pd.read_csv('cian_parsing_result_sale_1_100_sankt-peterburg_04_Apr_2023_studio.csv', 
                               encoding = 'windows-1251', sep = ';')
```

```python
spb_flat_1room_sale.loc[spb_flat_1room_sale.link == 'https://spb.cian.ru/sale/flat/280819981/','total_meters'] = 36.81
spb_flat_1room_sale['rooms_count'] = spb_flat_1room_sale['rooms_count'].replace(-1, 1)
spb_flat_2rooms_sale['rooms_count'] = spb_flat_2rooms_sale['rooms_count'].replace(-1,2)
spb_flat_3rooms_sale['rooms_count'] = spb_flat_3rooms_sale['rooms_count'].replace(-1,3)
spb_flat_4rooms_sale['rooms_count'] = spb_flat_4rooms_sale['rooms_count'].replace(-1,4)
spb_flat_5rooms_sale['rooms_count'] = spb_flat_5rooms_sale['rooms_count'].replace(-1,5)
spb_flat_studio_sale['rooms_count'] = 0.5
```

```python
msk_flat_1room_sale = pd.read_csv('cian_parsing_result_sale_1_100_moskva_10_Apr_2023_10_33_10_266174_1.csv', 
                                  encoding = 'windows-1251', sep = ';')
msk_flat_2rooms_sale = pd.read_csv('cian_parsing_result_sale_1_100_moskva_10_Apr_2023_14_54_42_931065_2.csv', 
                               encoding = 'windows-1251', sep = ';')

msk_flat_3rooms_part1 = pd.read_csv('cian_parsing_result_sale_1_100_moskva_10_Apr_2023_19_47_32_181691_3_part1.csv',
                               encoding = 'windows-1251', sep = ';')
msk_flat_3rooms_part2 = pd.read_csv('cian_parsing_result_sale_21_100_moskva_10_Apr_2023_20_53_51_987643_3_part2.csv',
                               encoding = 'windows-1251', sep = ';')
msk_flat_3rooms_part3 = pd.read_csv('cian_parsing_result_sale_26_100_moskva_10_Apr_2023_21_12_01_190962_3_part3.csv',
                               encoding = 'windows-1251', sep = ';')
msk_flat_3rooms_sale = pd.concat([msk_flat_3rooms_part1,msk_flat_3rooms_part2,msk_flat_3rooms_part3])


msk_flat_4rooms_sale = pd.read_csv('cian_parsing_result_sale_1_100_moskva_11_Apr_2023_12_56_58_475595_4.csv',
                               encoding = 'windows-1251', sep = ';')
msk_flat_5rooms_sale = pd.read_csv('cian_parsing_result_sale_1_100_moskva_11_Apr_2023_15_45_23_221731_5.csv',
                               encoding = 'windows-1251', sep = ';')

msk_flat_studio_part1 = pd.read_csv('cian_parsing_result_sale_1_100_moskva_10_Apr_2023_12_42_38_807183_studio_part1.csv', 
                               encoding = 'windows-1251', sep = ';')
msk_flat_studio_part2 = pd.read_csv('cian_parsing_result_sale_20_100_moskva_10_Apr_2023_13_26_09_548737_studio_part2.csv', 
                               encoding = 'windows-1251', sep = ';')
msk_flat_studio_sale = pd.concat([msk_flat_studio_part1,msk_flat_studio_part2])
```

```python
msk_flat_1room_sale['rooms_count'] = msk_flat_1room_sale['rooms_count'].replace(-1, 1)
msk_flat_2rooms_sale['rooms_count'] = msk_flat_2rooms_sale['rooms_count'].replace(-1,2)
msk_flat_3rooms_sale['rooms_count'] = msk_flat_3rooms_sale['rooms_count'].replace(-1,3)
msk_flat_4rooms_sale['rooms_count'] = msk_flat_4rooms_sale['rooms_count'].replace(-1,4)
msk_flat_5rooms_sale['rooms_count'] = msk_flat_5rooms_sale['rooms_count'].replace(-1,5)
msk_flat_studio_sale['rooms_count'] = 0.5
```

```python
data = pd.concat([spb_flat_1room_sale,
spb_flat_2rooms_sale,
spb_flat_3rooms_sale,
spb_flat_4rooms_sale,
spb_flat_5rooms_sale,
spb_flat_studio_sale,
msk_flat_1room_sale,
msk_flat_2rooms_sale,
msk_flat_3rooms_sale,
msk_flat_4rooms_sale,
msk_flat_5rooms_sale,
msk_flat_studio_sale])
```

```python
data
```

```python
data.to_csv('cian_data_sale.csv', sep = ';',index= False)
```
